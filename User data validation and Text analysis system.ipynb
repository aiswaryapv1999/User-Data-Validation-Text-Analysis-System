User Data Validation & Text Analysis System

You are developing a text processing module for an online application.
The system receives raw string inputs from users and must clean, validate,
and analyze the data before storing it.

```python
# Given Data:
user_data = [
 " Anu, anu@gmail.com , 21 ",
 "Ravi ,ravi123gmail.com, 17",
 " Suresh,suresh@gmail.com,25 ",
 "Meena,meena@gmail,22",
 ",invalid@gmail.com,30"
]
```

```python
user_data = [
    " Anu, anu@gmail.com , 21 ",
    "Ravi ,ravi123gmail.com, 17",
    " Suresh,suresh@gmail.com,25 ",
    "Meena,meena@gmail,22",
    ",invalid@gmail.com,30"
]

total_records = 0
valid_users = 0
invalid_users = 0
eligible_users = 0
underage_users = 0

domain_count = {}

for record in user_data:
    total_records += 1

    record = record.strip()
    details_of_user = record.split(",")

    # Must have exactly 3 fields
    if len(details_of_user) != 3:
        invalid_users += 1
        continue

    # Clean data
    cleaned_data = []
    for item in details_of_user:
        cleaned_data.append(item.strip())

    name = cleaned_data[0].title()
    email = cleaned_data[1].lower()
    age_str = cleaned_data[2].strip()

    valid_email = False
    under_age = False
    validation_errors = []

# AGE VALIDATION
    if age_str.isdigit():
        age = int(age_str)
        if age >= 18:
            eligible_users += 1
        else:
            under_age = True
            underage_users += 1
    else:
        age = 0
        under_age = True

# EMAIL VALIDATION, ALL DOMAIN COUNT

    if email.count("@") == 1 and email[0] != "@" and email[-1] != "@":
        domain = email.split("@")[1]
        if "." in domain:
            valid_email = True
            domain_count[domain] = domain_count.get(domain, 0) + 1
        else:
            validation_errors.append("Invalid email")
    else:
        validation_errors.append("Invalid email")

# NAME VALIDATION
    if name == "":
        validation_errors.append("Missing name")

    if under_age:
        validation_errors.append("Under age")

# FINAL VALID / INVALID CHECK
    if not validation_errors:
        valid_users += 1
        print("Valid user")
    else:
        invalid_users += 1
        print("Invalid user")

    output = f"Name: {name} | Email: {email} | Age: {age}"
    print(output)

    if validation_errors:
        print("Reasons:", validation_errors)
    print()

# SUMMARY REPORT
print("Summary Report")
print("--------------")
print("Total Records:", total_records)
print("Valid Users:", valid_users)
print("Invalid Users:", invalid_users)
print("Eligible Users:", eligible_users)
print("Underage Users:", underage_users)

print("\nDomain Summary:")
for domain, count in domain_count.items():
    print(domain, "-", count)
```
